**TALLER EXPLORANDO SHAPELY**

Shapely permite operaciones con ciertas entidades geométricas de las que se derivan las representaciones gráficas de elementos de interés en los mapas. Para esto emplea tres objetos geométricos fundamentales:
Puntos: son entidades definidas por dos coordenadas (x, y) sobre una superficie bidimensional. También se pueden definir puntos tridimensionales añadiendo una coordenada z.
Líneas: son colecciones de puntos interconectados, están delimitadas por dos puntos borde.
Polígonos: son colecciones de líneas que encierran un área.

Los puntos en Shapely vienen de la clase Point como se muestra en el código:

```python
from shapely.geometry import Point, LineString, Polygon
point = Point(1,1)

# Instalar las bibliotecas necesarias
!pip install shapely matplotlib

# Importar las bibliotecas
from shapely.geometry import Point, LineString
import matplotlib.pyplot as plt
```

Las líneas se crean utilizando la clase LineString y reciben una lista de puntos para su creación.


```python
#Creo 3 puntos que definan la linea

point1 = Point(0,0)
point2 = Point(3,1)
point3 = Point(0,3)

#Creo a LineString desde nuestros objetos Point
line1 = LineString([point1, point2, point3])

#También es posible utilizar conjuntos de coordenadas para crear directamente un objeto LineString.
line2 = LineString([(0,0), (3,1),(0,3)])

# Extraer las coordenadas de la línea
x, y = line1.xy

# Dibujar la línea
plt.plot(x, y, color='blue')
plt.scatter(*line1.xy, color='red')  # Dibujar puntos
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Línea definida por puntos')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

Y los polígonos se crean a partir de la clase Polygon, el cual puede iniciarse a partir de puntos o linestrings.

```python
#creamos 3 puntos que definen el contorno del poligono
point1 = Point(0,0)
point2 = Point(3,1)
point3 = Point(0,3)

polygon2= Polygon([[p.x, p.y] for p in [point1, point2, point3]])

# Extraer las coordenadas del polígono
x, y = polygon2.exterior.xy

# Dibujar el polígono
plt.plot(x, y, color='blue')
plt.fill(x, y, color='lightgray', alpha=0.5)  # Rellenar el polígono
plt.scatter(*polygon2.exterior.xy, color='red')  # Dibujar puntos
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono definido por puntos')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show
```

```python
#primero definimos nuetro exterior

exterior = Polygon([(0,0), (3,1),(0,3)])

#luego definimos la forme de el  agujero
hole = [[(0,1), (3,1),(0,1.5)]]

#Crear el polígono con el agujero
polygon3 = Polygon(shell = exterior, holes=hole)

# Extraer las coordenadas del polígono y el agujero
exterior_x, exterior_y = exterior.exterior.xy
hole_x, hole_y = polygon3.interiors[0].xy

# Dibujar el polígono y el agujero
plt.plot(exterior_x, exterior_y, color='blue')  # Dibujar el exterior del polígono
plt.fill(exterior_x, exterior_y, color='lightgray', alpha=0.5)  # Rellenar el exterior
plt.plot(hole_x, hole_y, color='red')  # Dibujar el agujero
plt.fill(hole_x, hole_y, color='white')  # Rellenar el agujero
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono con agujero')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```

```python
# Definir el exterior del polígono
exterior = Polygon([(0,0), (3,1), (0,3)])

# Definir el agujero del polígono
hole = [[(0,1), (3,1), (0,1.5)]]

# Crear el polígono con el agujero
polygon_with_hole = Polygon(shell=exterior.exterior, holes=hole)

# Extraer las coordenadas del polígono y el agujero
exterior_x, exterior_y = exterior.exterior.xy
hole_x, hole_y = polygon_with_hole.interiors[0].xy

# Dibujar el polígono y el agujero
plt.plot(exterior_x, exterior_y, color='blue')  # Dibujar el exterior del polígono
plt.fill(exterior_x, exterior_y, color='lightgray', alpha=0.5)  # Rellenar el exterior
plt.plot(hole_x, hole_y, color='red')  # Dibujar el agujero
plt.fill(hole_x, hole_y, color='white')  # Rellenar el agujero
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polígono con agujero')
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```
